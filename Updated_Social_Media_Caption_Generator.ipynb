```python
from openai import OpenAI
```

```python
# 🔑 Set your OpenAI API key
client = OpenAI(api_key="your-api-key-here")
```

```python
def generate_social_media_post(topic, platform="Instagram"):
    prompt = f"""
    Create a catchy and creative {platform} post based on the topic: \"{topic}\".

    Include:
    - An engaging caption
    - 5 relevant hashtags
    - Suitable emojis

    Output format:
    Caption: <Your Caption>
    Hashtags: <#hashtag1> <#hashtag2> ... 
    Emojis: <emoji1> <emoji2> ...
    """

    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.8,
        max_tokens=250
    )

    return response.choices[0].message.content
```

```python
topic = input("Enter your topic or idea: ")
platform = input("Which platform? (Instagram/Twitter/LinkedIn): ")

output = generate_social_media_post(topic, platform)
print("\n🔮 Generated Post:\n")
print(output)
```
